#data sanitization and data parsing

```python
df_new['sellerproductcount']= df_new['sellerproductcount'].str.split()
```

removing Chinese sellers from the data set according to the requirements

creating new columns as US seller and Germany seller to identify which seller is from which country

```python
df_new['US seller']=df_new[ 'businessaddress' ].str.contains('US')
df_new['Germany seller']=df_new[ 'businessaddress' ].str.contains('DE')
```

```python
df_new.head(10)
```

 parsing out phone numbers, and email IDs of merchants

```python
df_new['sellerdetails'] = df_new['sellerdetails'].replace(np.nan, '')
```

```python
df_new['sellerdetails'] = df_new['sellerdetails'].apply(str)
```

```python
def phad(x):
  pattern = "(\+\d{4}-\d{3}-\d{4})|([a-zA-Z0-9.-]+@[a-zA-Z0-9.-]+\.[a-zA-Z0-9_-]+)"
  ele = re.findall(pattern,x)
  op=[]
  for element in ele:
    for i in range (2):
        if (len(element[i])!=0):
            op.append(element[i])
  return op          
```

```python
df_new['ph em'] = df_new['sellerdetails'].apply(phad)
```

```python
#renaming the ph em column to phone_and_email
df_new.rename(index=str, columns={'ph em':'phone_and_email',
                                  'heroproduct1':'ratings_of_top_product1',
                                  'heroproduct2':'ratings_of_top_product2',
                                  'seller business name':'business_name',
                                  'Count of seller brands':'count_of_brands'
                                  }
                          , inplace=True)
```

```python
df_new.head(10)
```

```python
df_new['business_name']=df_new['business_name'].str.split(':')
```

```python
df_new['business_name']=df_new.business_name.str.slice(-1)
```

```python
df_new.business_name[0]
```

```python
df_new.head(10)
```

```python
#removing unwanted columns or information
#here the category for all products are Garden so it is unnessesary
df_new.drop('category', axis = 1, inplace = True)
```

```python
# here sellerlink-url ,sellerproductcount,sellerratings,brand_url , date_added are unnessesary so droping these columns
df_new.drop('sellerlink-url', axis = 1, inplace = True)
df_new.drop('sellerproductcount', axis = 1, inplace = True)
df_new.drop('sellerratings', axis = 1, inplace = True)
df_new.drop('brand_url', axis = 1, inplace = True)
df_new.drop('sellerstorefront-url', axis = 1, inplace = True)
df_new.drop('date_added', axis = 1, inplace = True)
```

```python
df_new.head(10)
```

```python
df_new['product_count']=df_new.product_count.apply(lambda x: ','.join(map(str, x)))
df_new['percofpositive_rating']=df_new.percofpositive_rating.apply(lambda x: ','.join(map(str, x)))
df_new['no_of_ratings']=df_new.no_of_ratings.apply(lambda x: ','.join(map(str, x)))
```

```python
df_new['percofpositive_rating']=df_new.percofpositive_rating.str.replace("%","")
df_new['no_of_ratings']=df_new.no_of_ratings.str.replace("(","")
```

```python
df_new['product_count'] = pd.to_numeric(df_new['product_count'].str.replace(',', ''), errors='coerce')
df_new['product_count']=df_new.product_count.astype(int)
```

```python
df_new['percofpositive_rating']=df_new.percofpositive_rating.astype(int)
```

```python
df_new['no_of_ratings'] = pd.to_numeric(df_new['no_of_ratings'].str.replace(',', ''), errors='coerce')
df_new['no_of_ratings']=df_new.no_of_ratings.astype(int,errors='ignore')
```

```python
df_new['top_product_rating']=df_new['ratings_of_top_product1']+df_new['ratings_of_top_product2']
```

```python
df_new.drop('ratings_of_top_product1', axis = 1, inplace = True)
df_new.drop('ratings_of_top_product2', axis = 1, inplace = True)
```
